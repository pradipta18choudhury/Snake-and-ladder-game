```python
import random

MAX_VAL = 100
DICE_FACE = 6

snakes={
    17:7,
    54:34,
    62:19,
    98:79
}

ladders={
    3:38,
    24:33,
    42:93,
    72:84
}


player_turn_text = [
    "Your turn.",
    "Go.",
    "Please proceed.",
    "I have to win this.",
    "Are you ready?",
    "",
]

snake_bite = [
    "boohoo",
    "bummer",
    "snake bite",
    "oh no",
    "dang",
    "Zingggg"
]

ladder_jump = [
    "woohoo",
    "woww",
    "nailed it",
    "oh my God...",
    "yaayyy",
    "I am about to win.."
]


def welcome_msg():
    msg = """
    Welcome to Snake and Ladder Game.


    Rules:
      1. Initally both the players are at starting position i.e. 0. 
         Take it in turns to roll the dice. 
         Move forward the number of spaces shown on the dice.
      2. If you lands at the bottom of a ladder, you can move up to the top of the ladder.
      3. If you lands on the head of a snake, you must slide down to the bottom of the snake.
      4. The first player to get to the FINAL position is the winner.

    """
    print(msg)


def get_player_names():
    player1_name = None
    while not player1_name:
        player1_name = input("Please enter a valid name for first player: ").strip()

    player2_name = None
    while not player2_name:
        player2_name = input("Please enter a valid name for second player: ").strip()

    print("\nMatch will be played between '" + player1_name + "' and '" + player2_name + "'\n")
    return player1_name, player2_name



def get_dice_value():
    dice_value = random.randint(1, DICE_FACE)
    print("Its a " + str(dice_value))
    return dice_value



def got_snake_bite(old_value, current_value, player_name):
    print("\n" + random.choice(snake_bite).upper() + " ~~>")
    print("\n" + player_name + " got a snake bite. Down from " + str(old_value) + " to " + str(current_value))


def got_ladder_jump(old_value, current_value, player_name):
    print("\n" + random.choice(ladder_jump).upper() + " ########")
    print("\n" + player_name + " climbed the ladder from " + str(old_value) + " to " + str(current_value))


def snake_ladder(player_name, current_value, dice_value):
    old_value = current_value
    current_value = current_value + dice_value

    if current_value > MAX_VAL:
        print("You need " + str(MAX_VAL - old_value) + " to win this game. Keep trying.")
        return old_value

    print("\n" + player_name + " moved from " + str(old_value) + " to " + str(current_value))
    if current_value in snakes:
        final_value = snakes.get(current_value)
        got_snake_bite(current_value, final_value, player_name)

    elif current_value in ladders:
        final_value = ladders.get(current_value)
        got_ladder_jump(current_value, final_value, player_name)

    else:
        final_value = current_value

    return final_value



def check_win(player_name, position):
    if MAX_VAL == position:
        print("\n\n\nThats it.\n\n" + player_name + " won the game.")
        print("Congratulations " + player_name)
        print("\nThank you for playing the game.\n")
        exit(0)


def start():
    welcome_msg()
    player1_name, player2_name = get_player_names()
    player1_current_position = 0
    player2_current_position = 0

    while True:
        input_1 = input("\n" + player1_name + ": " + random.choice(player_turn_text) + " type roll or enter digit (1 - 20) to play your side: ")
        #temp=int(input_1)
        if input_1=="exit" :
            print("congratulations!!! " + player2_name + " is the winner..")
            print("\n\n thanks for playing\n\n")
            print("##########Game successfully finished##########\n\n")
            break
        # this is for auto mode of the game
        elif input_1=="roll" :
            print("\nRolling dice...")
            dice_value = get_dice_value()
            print(player1_name + " moving....")
            player1_current_position = snake_ladder(player1_name, player1_current_position, dice_value)
            check_win(player1_name, player1_current_position)
        # this is for manual mode of the game        
        #elif isinstance(input_1,int) == True and int(input_1)>=1 and int(input_1)<=20 :
        #elif type(input_1) == 'int' and int(input_1)>=1 and int(input_1)<=20 :
        elif input_1.isdigit() == True and int(input_1)>=1 and int(input_1)<=20 :
            print("\n Got the number...")
            dice_value=int(input_1)
            print(player1_name + " moving....")
            player1_current_position = snake_ladder(player1_name, player1_current_position, dice_value)
            check_win(player1_name, player1_current_position)
        elif input_1 != "roll" or input_1 != "exit" or type(int(input_1)) != int :
            print(" \n you lost your choice...")
            print("\n Wait for your next turn..")
        else :
            continue
            
        
        
        input_2 = input("\n" + player2_name + ": " + random.choice(player_turn_text) + " type roll or enter digits (1 - 20) to play your side: ")
        if input_2=="exit" :
            print("congratulations!!! " + player1_name + " is the winner..")
            print("\n\n Thanks for playing\n")
            print("##########  Game successfully finished  ##########\n\n")
            break
        elif input_2=="roll" :
            print("\nRolling dice...")
            dice_value = get_dice_value()
            print(player2_name + " moving....")
            player2_current_position = snake_ladder(player2_name, player2_current_position, dice_value)
            check_win(player2_name, player2_current_position)
        elif input_2.isdigit() == True and int(input_1)>=1 and int(input_1)<=20 :
            print("\n Got the number...")
            dice_value=int(input_2)
            print(player2_name + " moving....")
            player2_current_position = snake_ladder(player2_name, player2_current_position, dice_value)
            check_win(player2_name, player2_current_position)
        elif input_2 != "roll" or input_2 != "exit" or type(int(input_2)) > 20 :
            print(" \n you lost your choice...")
            print("\n Wait for your next turn\n\n")
        else :
            continue
            
        

if __name__ == "__main__":
    start()
```
